```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
df['Species'].value_counts()
```

```python
sns.heatmap(df.corr())
```

```python
df_setosa = df[df['Species']=='Iris-setosa']
df_virginica=df[df['Species']== 'Iris-virginica']
df_versicolor=df[df['Species']== 'Iris-versicolor']
```

```python
plt.plot(df_setosa['PetalLengthCm'],np.zeros_like(df_setosa['PetalLengthCm']),'o')
plt.plot(df_virginica['PetalLengthCm'],np.zeros_like(df_virginica['PetalLengthCm']),'o')
plt.plot(df_versicolor['PetalLengthCm'],np.zeros_like(df_versicolor['PetalLengthCm']),'o')
plt.xlabel('PetalLength')
```

```python
sns.FacetGrid(df,hue ='Species',size =5).map(plt.scatter,"PetalLengthCm","SepalWidthCm").add_legend()
```

```python
sns.pairplot(df,hue='Species',size=5)
```
